```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/stats.csv')
```

```python
data.head()
```

```python
df = data.loc[:, ['team', 'wins', 'losses', 'season']]
df['draws'] = 38 - df['wins'] - df['losses']
```

```python
df['points'] = df['wins'] * 3 + df['draws'] * 1
df['points'] = df['points'].astype(int)
```

```python
desired_columns_order = ['team', 'wins', 'losses', 'draws', 'points', 'season']
df = df[desired_columns_order]
```

```python
df.head()
```

```python
for season, group in df.groupby("season"):
    season_csv_filename = f'points_season_{season}.csv'
    group.to_csv("../data/" + season_csv_filename, index=False)
    print(f"Saved {season_csv_filename}")
```

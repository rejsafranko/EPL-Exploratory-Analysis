```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/stats.csv')
data.head()
```

```python
data.columns
```

```python
filtered_df = data.loc[:, ['team', 'ontarget_scoring_att', 'att_hd_goal', 'att_pen_goal',  'att_freekick_goal', 
                           'att_ibox_goal', 'att_obox_goal', 'goal_fastbreak', 'total_offside', 'season']]
```

```python
filtered_df.head()
```

```python
for season, group in filtered_df.groupby("season"):
    season_csv_filename = f'attacking_stats_{season}.csv'
    group.to_csv("../data/" + season_csv_filename, index=False)
    print(f"Saved {season_csv_filename}")
```

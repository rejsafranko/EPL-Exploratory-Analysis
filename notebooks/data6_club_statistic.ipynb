```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/stats.csv')
```

```python
data.head()
```

```python
filtered_df = data.loc[:, ['team', 'wins', 'losses', 'total_yel_card', 'total_red_card', 'hit_woodwork', 'goals', 'goals_conceded', 'clean_sheet', 'season']]
```

```python
filtered_df
```

```python
for season, group in filtered_df.groupby("season"):
    season_csv_filename = f'general_stats_{season}.csv'
    group.to_csv("../data/" + season_csv_filename, index=False)
    print(f"Saved {season_csv_filename}")
```
